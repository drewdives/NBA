```python
import pandas as pd
data_filename = "https://raw.githubusercontent.com/drewdives/datasets-diabetes/main/NBA22_23"
dataset = pd.read_csv(data_filename)
```

```python
# for now - drop arena and attendance (later - you can add this back as part of feature engineering)
dataset.drop(["Arena","Attend."],axis=1, inplace=True)
```

```python
dataset.columns = ["Date", "Start (ET)", "Visitor Team", "VisitorPts","Home Team", "HomePts", "OT?", "Score Type", "Notes"]
```

```python
dataset.head()
```

```python
dataset["HomeWin"] = dataset["VisitorPts"] < dataset["HomePts"]
dataset.head()
```

```python
y_true = dataset["HomeWin"].values
y_true
```

```python
dataset["HomeWin"].mean()
```

```python
from collections import defaultdict
won_last = defaultdict(int)
```

```python
dataset["HomeLastWin"] = 0
dataset["VisitorLastWin"] = 0
```

```python
dataset.head()
```

```python
for index, row in dataset.iterrows():
    home_team = row["Home Team"]
    visitor_team = row["Visitor Team"]
    row["HomeLastWin"] = won_last[home_team]
    dataset.at[index, "HomeLastWin"] = won_last[home_team]
    dataset.at[index, "VisitorLastWin"] = won_last[visitor_team]
    won_last[home_team] = int(row["HomeWin"])
    won_last[visitor_team] = 1 - int(row["HomeWin"])
```

```python
import os
standings_filename = "https://raw.githubusercontent.com/drewdives/datasets-diabetes/main/2022-23%20NBA%20Standings"
standings = pd.read_csv(standings_filename, skiprows=1)
standings.head()
```

```python
dataset["HomeTeamRanksHigher"] = 0
for index, row in dataset.iterrows():
    home_team = row["Home Team"]
    visitor_team = row["Visitor Team"]
    home_rank = standings[standings["Team"] == home_team]["Rk"].values[0]
    visitor_rank = standings[standings["Team"] == visitor_team]["Rk"].values[0]

    # Set the value using .at
    dataset.at[index, "HomeTeamRanksHigher"] = int(home_rank < visitor_rank)
dataset.head()
```

```python
X_homehigher = dataset[["HomeLastWin", "VisitorLastWin", "HomeTeamRanksHigher"]].values
```

```python
clf = DecisionTreeClassifier(random_state=14)
scores = cross_val_score(clf, X_homehigher, y_true, scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
from collections import defaultdict

last_match_winner = defaultdict(int)
dataset["HomeTeamWonLast"] = 0

for index, row in dataset.iterrows():
    home_team = row["Home Team"]
    visitor_team = row["Visitor Team"]
    teams = tuple(sorted([home_team, visitor_team]))  # Sort for a consistent ordering

    # Set in the row, who won the last encounter
    home_team_won_last = 1 if last_match_winner[teams] == home_team else 0
    dataset.at[index, "HomeTeamWonLast"] = home_team_won_last

    # Who won this one?
    winner = home_team if row["HomeWin"] else visitor_team
    last_match_winner[teams] = winner
```

```python
import pandas as pd

# Load the dataset
file_path = 'https://github.com/fenago/datasets/raw/main/NBA2015_2016.txt'
dataset = pd.read_csv(file_path)

# Preprocess the dataset
dataset.drop(["Arena", "Attend."], axis=1, inplace=True)
dataset["Date"] = pd.to_datetime(dataset["Date"])
dataset.columns = ["Date", "Start (ET)", "Visitor Team", "VisitorPts", "Home Team", "HomePts", "OT?", "Score Type", "Notes"]

# Sort the dataset by date
dataset.sort_values(by="Date", inplace=True)

# Initialize a dictionary to track game outcomes for each team
team_games = {}

# Populate the dictionary with game outcomes
for index, row in dataset.iterrows():
    home_team = row["Home Team"]
    visitor_team = row["Visitor Team"]
    home_win = row["HomePts"] > row["VisitorPts"]

    if home_team not in team_games:
        team_games[home_team] = []
    if visitor_team not in team_games:
        team_games[visitor_team] = []

    team_games[home_team].append(home_win)
    team_games[visitor_team].append(not home_win)

# Function to count wins in the last 10 games
def count_recent_wins(team_name):
    if len(team_games[team_name]) < 10:
        return sum(team_games[team_name])
    else:
        return sum(team_games[team_name][-10:])

# Apply the function to each row in the dataset
for index, row in dataset.iterrows():
    home_team = row["Home Team"]
    visitor_team = row["Visitor Team"]

    dataset.at[index, "HomeTeamWinsLast10"] = count_recent_wins(home_team)
    dataset.at[index, "VisitorTeamWinsLast10"] = count_recent_wins(visitor_team)

# Display the dataset with the new feature
print(dataset[["Date", "Home Team", "Visitor Team", "HomeTeamWinsLast10", "VisitorTeamWinsLast10"]].head())
```
